```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Depot is at origin, all angles are calculated from depot

```python
df = pd.read_csv('/content/cVRPsTW250_1.csv')
```

```python
df
```

```python
Angle = []
for i in df.index:
  angle = math.atan2(df.loc[i].at['LON'],df.loc[i].at['LAT'])*180/np.pi 
  if angle < 0:
    angle = angle + 360
  Angle.append(angle)
```

```python
# Angle
```

```python
df['theta'] = Angle  #in degrees
```

```python
# Sorting by column 'theta'
ndf = df.sort_values(by=['theta']).reset_index(drop=True)
```

```python
ndf
```

```python
Angle = list(set(Angle))
```

```python
Angle.sort()
```

```python
# Angle
```

```python
uniq  = Angle.copy()   # to get the max dif angle from neighbors
# uniq
```

```python
uniq.insert(0,uniq[len(uniq)-1])
uniq.append(uniq[1])
```

```python
# uniq
```

```python
max_dif = []
for i in range(1,len(uniq)-1):
  dif = uniq[i+1] - uniq[i-1]
  if dif < 0 :
    dif = dif + 360
  max_dif.append(dif)
# max_dif
```

```python
index = max_dif.index(max(max_dif))
index  # starting demand point from depot
```

```python
ang=Angle[index]
ang
```

```python
ndf
```

```python
ndf[ndf['theta']==ang]
```

```python
#to get index of ang
index = ndf[ndf['theta']==ang].index.values[0]
```

```python
covered_theta = list(ndf['theta'][index:])
covered_q = list(ndf['ORDERWEIGHT'][index:])
covered_i = list(ndf['ORDERID'][index:])
```

```python
rem_theta = list(ndf['theta'][:index])
rem_q = list(ndf['ORDERWEIGHT'][:index])
rem_i = list(ndf['ORDERID'][:index])
```

```python
new_theta = covered_theta + rem_theta
new_q = covered_q + rem_q
new_i = covered_i + rem_i
```

```python
dic = {
    'i': new_i,
    'q': new_q,
    'theta': new_theta
}
```

```python
# Anticloockwise direction
odf = pd.DataFrame(dic)
odf
```

```python
C= 50
```

```python
cluster = dict()
cluster.update({1:[]})
```

```python
Q = list(odf['q'])
```

```python
sum =0
j=1
for i in range(odf.shape[0]):
  sum += odf.loc[i,'q']
  if sum < C:
    cluster[j].append(odf.loc[i,'i'])
  else:
    sum = odf.loc[i,'q']
    j+=1
    cluster.update({j:[]})
    cluster[j].append(odf.loc[i,'i'])
```

```python
for r in cluster.values():
  r.insert(0,0)
```

```python
# Anticlockwise Direction
cluster
```
